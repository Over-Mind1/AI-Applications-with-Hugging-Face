```python
from transformers import pipeline
```

```python
text2textblender_bot= pipeline(task="text2text-generation",
                       model="facebook/blenderbot-400M-distill")
```

```python
user_message="Hello, how are you?"
text2textblender_bot.tokenizer(user_message)
```

```python
conversation = []
bot_response=text2textblender_bot(user_message)
conversation.append(("User", user_message))
conversation.append(("Bot", bot_response[0]['generated_text']))
print(conversation)
```

```python
conversation = []
while True:
    user_message = input("You: ")
    if not user_message:
        break
    conversation.append(("User", user_message))
    history = " ".join(turn for _, turn in conversation)
    bot_response = text2textblender_bot(history)[0]["generated_text"]
    conversation.append(("Bot", bot_response))
    print("Bot:", bot_response)
```
